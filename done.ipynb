```python
from sqlalchemy import create_engine
import pandas as pd
import getpass
passw = getpass.getpass()

engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                    .format(user="root",
                            pw=passw,
                            db=f'sakila'))


```

```python
def rentals_month(engine,year,month): 
    query = (f'''SELECT * FROM sakila.rental 
             WHERE rental_date LIKE "{year}-_{month}%%";''')
    my_df = pd.read_sql_query(query, engine)# pd.read_sql_query gives it with columns name 
    return my_df
```

```python
my_df = rentals_month(engine,2005,5)
my_df2 = rentals_month(engine,2005,6)
```

```python
def rentals_month(df,month,year):
    rentals = (f"rentals_{month}_{year}")
    i = pd.DataFrame(df.groupby(["customer_id"]).agg(rentalss=("customer_id", "count"))).reset_index()
    i = i.rename(columns= {"rentalss": rentals})
    return i
```

```python
my_df = rentals_month(my_df,5,2005)
my_df2 = rentals_month(my_df2,6,2005)
```
