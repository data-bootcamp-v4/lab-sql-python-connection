```python
from sqlalchemy import create_engine
```

```python
my_engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                        .format(user="root",
                               pw = getpass(),
                               db= "sakila"))
```

```python
def rentals_month(engine, month, year):
    
    my_engine = engine.connect()

    query = f"""SELECT *
                FROM rental
                WHERE EXTRACT(MONTH FROM rental_date) = {month} AND EXTRACT(YEAR FROM rental_date) = {year}"""

    
    rental_df = pd.read_sql(query, my_engine)
    
    return rental_df

rental_data = rentals_month(my_engine, month=5, year=2005)
```
